* Columns Normalisation :
  I observed that columns are in string format so i  have to convert themso i can use them in analysis 

```python
#1) Genres column:
movies.iloc[0].genres
```

```python
import ast
ast.literal_eval
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
#2) Keywords column:
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
#3) Cast columns:
    # for cast column i have to get names of three lead cast so I will create function based on above function for taht.
```

```python
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    
    return L    
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
#4) Crew column:
   #in this i obsverved that there are multiple dictionaries soi have extract the director name from job dictionary.So i will create new function based on above filter
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L        
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
#5) overview column: this is string column now i will convert it in list
```

```python
movies['overview']= movies['overview'].apply(lambda x:x.split())
```

* In this dataset we have to remove space so that same names/words does not clash with other words which can cause confusion. 

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

* In tags column i see that some words are repeating so i have to handle that.

```python
import nltk 
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

* Text Vectorisation :

```python
from sklearn.feature_extraction.text  import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words= 'english')
```

```python
vectors= cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index= new_df[new_df['title']==movie].index[0]
    distances= similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Batman Begins')
```
